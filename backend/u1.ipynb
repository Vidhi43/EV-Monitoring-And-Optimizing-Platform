```python
import random
import pandas as pd
from datetime import datetime, timedelta
```

```python
station_coords = [
    (18.48935, 74.04160),
    (18.49444, 73.99874),
    (18.48996, 74.03606),
    (18.49012, 74.04731),
    (18.51774, 73.97399)
]
```

```python
days = 30
price_per_kwh = 15
avg_kwh_per_session = 20
```

```python
for i, (lat, lon) in enumerate(station_coords, start=101):
    data = []
    date = datetime.now() - timedelta(days=days)

    for _ in range(days):
        voltage = round(random.uniform(210, 240), 2)
        current = round(random.uniform(10, 32), 2)
        temperature = round(random.uniform(25, 60), 2)
        leak_detected = random.choice([0, 0, 0, 1])  # 25% chance
        charging_sessions = random.randint(20, 100)
        faults = random.choice([0, 0, 1])  # 33% chance
        revenue = charging_sessions * avg_kwh_per_session * price_per_kwh

        data.append({
            "station_id": i,
            "date": date.strftime("%Y-%m-%d"),
            "voltage": voltage,
            "current": current,
            "temperature": temperature,
            "leak_detected": leak_detected,
            "charging_sessions": charging_sessions,
            "fault_occurred": faults,
            "revenue_inr": revenue,
            "latitude": lat,
            "longitude": lon
        })

        date += timedelta(days=1)

    # Convert to DataFrame and save
    df = pd.DataFrame(data)
    filename = f"ev_station_{i}_daily_data.csv"
    df.to_csv(filename, index=False)

    print(f"✅ Dataset created successfully: {filename}")
    print(df.head(5))
```
